```
photo="/content/drive/MyDrive/face_detection/photos/Ekran görüntüsü 2022-09-18 223839.png"

r=cv2.imread(photo)
gray_photo=cv2.cvtColor(r, cv2.COLOR_BGR2RGB) # kanal sayısının azaltılması için yaptığımız bir işlem.

key=cv2.waitKey(0) # 0 diyerek sınırsız süre boyunca klavyeden bır karakter gırılene kadar bekle

cv2_imshow(gray_photo)


"""
if key == ord('q'):
  break
"""
```

```
gray_photo.shape
```

```
faces=face_cascade.detectMultiScale(gray_photo)
faces
```

```
for (x, y, w, h) in faces:
    cv2.rectangle(gray_photo, (x, y), (x+w, y+h), (0, 255, 0), 2)
cv2_imshow(gray_photo)
```
