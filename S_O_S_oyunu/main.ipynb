```python

class Play():

  def __init__(self):
    self.Oyun_Durumu=True
    self.Tahta=[['*', '*', '*', '*'],['O', '*', '*', 'S'],['S', '*', '*', 'S'],['O', '*', '*', 'O'],['*', 'O', 'S', '*']]
    self.sıra=0
    self.puan1=0
    self.puan2=0
    self.liste=[]



  def Tahtaa(self):
    for m in self.Tahta:
      for n in m:
        print(n, end=" ")
      print("\n")



  def Tablo(self):
    for v1 in self.Tahta:
      for v2 in v1:
        if v2=="*":
          self.liste.append("*")
        else:
          pass
    if len(self.liste)>0:
      self.sıraa()
    else:
      self.son()



  def girisler(self):
    while self.Oyun_Durumu:
      try:
        satır=int(input("satır= "))
        sütun=int(input("sütun= "))
        if 0<=satır<=len(self.Tahta) and 0<=sütun<=len(self.Tahta[0]):
          while True:
            try:
              print("Girmek istediğiniz harf S/O")
              harf=input().upper()
              if harf=="S" or harf=="O":
                self.Dolu_Mu(satır,sütun,harf)
              else:
                raise
              break
            except:
              print("uygun olmayan giriş!!! lütfen S/O giriniz")

        else:
          print("satır ve sütun sadece 0-2 aralığında olabilir.")
          self.girisler()

      except:
        print("lütfen numerıc deger gırınız")


  def Player1(self):
    print("sıra sizde Player1")
    self.reset()
    self.girisler()


  def Player2(self):
    print("sıra sizde Player2")
    self.reset()
    self.girisler()


  def sıraa(self):
    if self.Oyun_Durumu==True:
      self.Tahtaa()
      if self.sıra%2==0:
        self.Player1()

      else:
        self.Player2()


  def son(self):
    if self.puan1>self.puan2:
      print(f"oyunu {self.puan1} puan ile kazandınız player2 ")
      self.Close()

    elif self.puan1==self.puan2:
      print(f"oyun {self.puan1}-{self.puan2} ile berabere bitti. ")
      self.Close()

    else:
      print(f"oyunu {self.puan2} puan ile kazandınız player2 ")
      self.Close()



  def reset(self):
    (self.liste).clear()



  def puan(self):
    if self.sıra%2==0:
      self.puan1=(self.puan1)+1
      self.Tablo()
    else:
      self.puan2=(self.puan2)+1
      self.Tablo()


  def Dolu_Mu(self,satır,sütun,harf):
    if self.Tahta[satır][sütun]=="*":
      self.Tahta[satır][sütun]=harf
      self.Tahta_Kontrol(satır,sütun)

    else:
      print("dolu koordinat lütfen farklı bir konum seçiniz..")
      self.girisler()



  def Tahta_Kontrol(self, satır, sütun):
    try:
        if (
                (sütun - 1 >= 0 and sütun + 1 < len(self.Tahta[satır]) and
                  (self.Tahta[satır][sütun - 1] + self.Tahta[satır][sütun] + self.Tahta[satır][sütun + 1]) == "SOS") or
                (satır - 1 >= 0 and sütun - 1 >= 0 and satır + 1 < len(self.Tahta) and sütun + 1 < len(
                    self.Tahta[satır]) and
                  (self.Tahta[satır - 1][sütun - 1] + self.Tahta[satır][sütun] + self.Tahta[satır + 1][
                      sütun + 1]) == "SOS") or
                (satır - 1 >= 0 and sütun + 1 < len(self.Tahta[satır]) and satır + 1 < len(self.Tahta) and
                  sütun - 1 >= 0 and
                  (self.Tahta[satır - 1][sütun + 1] + self.Tahta[satır][sütun] + self.Tahta[satır + 1][
                      sütun - 1]) == "SOS") or
                (satır - 1 >= 0 and satır + 1 < len(self.Tahta) and
                  (self.Tahta[satır - 1][sütun] + self.Tahta[satır][sütun] + self.Tahta[satır + 1][sütun]) == "SOS") or
                (satır - 2 >= 0 and
                  (self.Tahta[satır][sütun] + self.Tahta[satır - 1][sütun] + self.Tahta[satır - 2][sütun]) == "SOS") or
                (satır + 2 < len(self.Tahta) and
                  (self.Tahta[satır][sütun] + self.Tahta[satır + 1][sütun] + self.Tahta[satır + 2][sütun]) == "SOS")
        ):
            self.puan()
        else:
            self.sıra += 1
            self.Tablo()

    except :
        pass

  def Close(self):
    print("Oyun sonlanmıstır..")
    self.Oyun_Durumu=False



play=Play()
while play.Oyun_Durumu:
  play.sıraa()
```
