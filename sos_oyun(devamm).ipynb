```
class Play():

  def __init__(self):
    self.Oyun_Durumu=True
    self.Tahta=[['S', 'O', 'O', 'S'],['O', '*', 'S', 'S'],['O', 'S', 'S', 'S'],['O', 'O', '*', 'S'],['S', 'O', 'S', 'S']]
    self.sıra=0
    self.puan1=0
    self.puan1=0
    self.liste=[]



  def Tahtaa(self):
    for m in self.Tahta:
      for n in m:
        print(n, end=" ")
      print("\n")



  def Tablo(self):
    for v1 in self.Tahta:
      for v2 in v1:
        if v2=="*":
          self.liste.append("*")
        else:
          pass
    if len(self.liste)>0:
      self.sıraa()
    else:
      self.Close()



  def girisler(self):
    while self.Oyun_Durumu:
      try:
        satır=int(input("satır= "))
        sütun=int(input("sütun= "))
        if 0<=satır<=len(self.Tahta) and 0<=sütun<=len(self.Tahta[0]):
          while True:
            try:
              print("Girmek istediğiniz harf S/O")
              harf=input()
              if harf=="S" or harf=="O":
                self.Dolu_Mu(satır,sütun,harf)
              else:
                raise
              break
            except:
              print("uygun olmayan giriş!!! lütfen S/O giriniz")

        else:
          print("satır ve sütun sadece 1-3 aralığında olabilir.")
          self.girisler()

      except:
        print("lütfen numerıc deger gırınız")


  def Player1(self):
    print("sıra sizde Player1")
    self.girisler()


  def Player2(self):
    print("sıra sizde Player2")
    self.girisler()


  def sıraa(self):
    if self.Oyun_Durumu==True:
      self.Tahtaa()
      if self.sıra%2==0:
        self.Player1()

      elif self.sıra% 2 != 0:
        self.Player2()


  def son(self):
    if self.puan1>self.puan2:
      print("oyunu kazandınız player1 ")
      self.Close()

    elif self.puan1==self.puan2:
      print("oyun berabere bitti. ")

    else:
      print("oyunu kazandınız player2 ")
      self.Close()


  def puan(self):
    if self.sıra%2==0:
      self.puan1=self.puan1+1
      self.Tablo()
    else:
      self.puan2=self.puan2+1
      self.Tablo()


  def Dolu_Mu(self,satır,sütun,harf):
    if self.Tahta[satır][sütun]=="*":
      self.Tahta[satır][sütun]=harf
      self.Tahta_Durumu(satır,sütun)
      self.sıra+=1

    else:
      print("dolu koordinat lütfen farklı bir konum seçiniz..")
      self.girisler()



  def Tahta_Durumu(self,satır,sütun):

    if (((self.Tahta[satır][sütun-1] + self.Tahta[satır][sütun] + self.Tahta[satır][sütun+1])=="SOS") or
    ((self.Tahta[satır-1][sütun-1] + self.Tahta[satır][sütun] + self.Tahta[satır+1][sütun+1])=="SOS") or
    ((self.Tahta[satır-1][sütun+1] + self.Tahta[satır][sütun] + self.Tahta[satır+1][sütun-1])=="SOS") or
    ((self.Tahta[satır-1][sütun] + self.Tahta[satır][sütun] + self.Tahta[satır+1][sütun])=="SOS")):

      self.puan()
      self.Tablo()



    else:
      self.Tablo()


  def Close(self):
    print("Oyun sonlanmıstır..")
    self.Oyun_Durumu=False



play=Play()
while play.Oyun_Durumu:
  play.sıraa()
```
